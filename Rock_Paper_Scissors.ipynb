```python
import random
import os

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "scissors" and computer_choice == "paper") or \
         (user_choice == "paper" and computer_choice == "rock"):
        return "user"
    else:
        return "computer"

def clear_screen():
    # Clear the terminal screen for better visuals
    os.system('cls' if os.name == 'nt' else 'clear')

def display_header():
    print("=" * 40)
    print("ğŸŒŸ Welcome to Rock, Paper, Scissors! ğŸŒŸ".center(40))
    print("=" * 40)

def display_choices(user_choice, computer_choice):
    print("\nYour Choice:      ğŸ®", user_choice.capitalize())
    print("Computer Choice:  ğŸ¤–", computer_choice.capitalize())

def display_scores(user_score, computer_score, ties):
    print("\nğŸ“Š Current Scores ğŸ“Š")
    print("-" * 40)
    print(f"ğŸ® You: {user_score}".ljust(20) + f"ğŸ¤– Computer: {computer_score}".rjust(20))
    print(f"ğŸ¤ Ties: {ties}".center(40))
    print("-" * 40)

def display_result(result):
    if result == "tie":
        print("\nğŸ¤ It's a Tie!")
    elif result == "user":
        print("\nğŸ‰ You Win this Round!")
    else:
        print("\nğŸ˜” Computer Wins this Round!")

def rock_paper_scissors():
    user_score = 0
    computer_score = 0
    ties = 0

    while True:
        clear_screen()
        display_header()
        display_scores(user_score, computer_score, ties)

        print("\nChoose Your Option:")
        print("1. Rock ğŸª¨")
        print("2. Paper ğŸ“„")
        print("3. Scissors âœ‚ï¸")
        print("4. Exit ğŸšª")

        try:
            choice = int(input("\nEnter your choice (1-4): ").strip())
            if choice == 4:
                clear_screen()
                display_header()
                print("\nThanks for playing! ğŸ‘‹ Final Scores:")
                display_scores(user_score, computer_score, ties)
                break

            if choice not in [1, 2, 3]:
                print("\nâš ï¸ Invalid Choice. Please select a valid option.")
                input("\nPress Enter to try again...")
                continue

            user_choice = ["rock", "paper", "scissors"][choice - 1]
            computer_choice = get_computer_choice()

            clear_screen()
            display_header()
            display_choices(user_choice, computer_choice)

            result = determine_winner(user_choice, computer_choice)
            if result == "tie":
                ties += 1
            elif result == "user":
                user_score += 1
            else:
                computer_score += 1

            display_result(result)
            input("\nPress Enter to continue...")

        except ValueError:
            print("\nâš ï¸ Invalid Input. Please enter a number between 1 and 4.")
            input("\nPress Enter to try again...")

if __name__ == "__main__":
    rock_paper_scissors()
```
